```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
import re

```

```python
file_path_v1 = '/Users/nesta99/Documents/Nesta Personal/Personal/UNI/Machine Learning/ML/ZeroAttack/dataset/zero_day_attack_detection_dataset_V1-400k.csv'
file_path_v2 = '/Users/nesta99/Documents/Nesta Personal/Personal/UNI/Machine Learning/ML/ZeroAttack/dataset/zero_day_attack_detection_dataset_V2_97k.csv'
```

```python
df1 = pd.read_csv(file_path_v1)
df2 = pd.read_csv(file_path_v2)
```

```python
# Step 1: Check for missing values and handle them (fill or drop)
missing_v1 = df1.isnull().sum()
missing_v2 = df2.isnull().sum()
```

```python
# Print columns with missing values (if any)
print("Missing values in V1 dataset:")
print(missing_v1[missing_v1 > 0])
```

```python
print("\nMissing values in V2 dataset:")
print(missing_v2[missing_v2 > 0])
```

```python
# Concatenate the two datasets
df = pd.concat([df1, df2], ignore_index=True)
```

```python
# Display the first few rows
print(df.head())

# Get basic information
print(df.info())

# Check for missing values
print(df.isnull().sum())
```

```python
# Drop rows with missing values (if any)
df = df.dropna()
```

```python
# List of categorical columns
categorical_cols = ['Protocol', 'Flag', 'Family', 'Clusters', 'Application Layer Data', 'User-Agent', 'Geolocation', 'Event Description', 'Error Code']

# Initialize LabelEncoder
le = LabelEncoder()

# Encode categorical columns
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
```

```python
# Define feature columns
feature_cols = ['Protocol', 'Flag', 'Netflow Bytes', 'Port', 'Payload Size', 'Number of Packets', 'Response Time', 'Data Transfer Rate']

# Define the target variable
target_col = 'Threat Level'
```

```python
# Encode the target variable
df[target_col] = le.fit_transform(df[target_col])
```

```python
def extract_numeric(value):
    if isinstance(value, str):
        # Use regular expression to find numeric part
        match = re.search(r'[\d\.]+', value)
        if match:
            return float(match.group())
        else:
            return np.nan  # or 0.0 or any default value you prefer
    else:
        return value
```

```python
# Apply the function to 'Response Time'
df['Response Time'] = df['Response Time'].apply(extract_numeric)

# Apply the function to 'Data Transfer Rate'
df['Data Transfer Rate'] = df['Data Transfer Rate'].apply(extract_numeric)
```

```python
# Check for missing values after extraction
print(df[feature_cols].isnull().sum())

# Optionally, fill missing values or drop rows
df = df.dropna(subset=feature_cols)
```

```python
# Initialize StandardScaler
scaler = StandardScaler()

# Standardize the features
df[feature_cols] = scaler.fit_transform(df[feature_cols])
```

```python
# Split the data
X = df[feature_cols]
y = df[target_col]

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# Train the model
rf_classifier = RandomForestClassifier(n_estimators=100, random_state=42)
rf_classifier.fit(X_train, y_train)
```

```python
# Evaluate the model
y_pred = rf_classifier.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(
    cm, annot=True, fmt='d', cmap='Blues',
    xticklabels=le.classes_, yticklabels=le.classes_
)
plt.ylabel('Actual')
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.show()
```

```python
import joblib
import tensorflow as tf

best_rf_classifier.fit(X_train, y_train)

# Save the trained model as a .pkl file
joblib.dump(best_rf_classifier, 'network_traffic_classifier.pkl')

# Load the saved scikit-learn Random Forest model
loaded_rf_model = joblib.load('network_traffic_classifier.pkl')

# Define a function to make a Keras model that is functionally equivalent
def scikit_learn_to_keras(rf_model):
    inputs = tf.keras.Input(shape=(X_train.shape[1],))

    # A Lambda layer to use the sklearn model's predict_proba function
    def predict_proba_with_rf(inputs):
        np_inputs = inputs.numpy()
        rf_probs = rf_model.predict_proba(np_inputs)
        return rf_probs

    # Wrap predict_proba function using a Lambda layer
    outputs = tf.keras.layers.Lambda(
        lambda x: tf.py_function(func=predict_proba_with_rf, inp=[x], Tout=tf.float32),
        output_shape=(rf_model.n_classes_,)
    )(inputs)

    # Create the equivalent Keras model
    keras_model = tf.keras.Model(inputs=inputs, outputs=outputs)
    return keras_model

# Convert scikit-learn Random Forest model to Keras model
tf_keras_model = scikit_learn_to_keras(loaded_rf_model)

# Save the Keras model as a TensorFlow SavedModel
# Removed deprecated save_format argument
tf_keras_model.save('output/network_traffic_classifier_saved_model.keras')

print("Model has been successfully converted and saved as TensorFlow SavedModel format.")
```

```python
import joblib
import tensorflow as tf
import numpy as np

# Load the saved scikit-learn Random Forest model
loaded_rf_model = joblib.load('network_traffic_classifier.pkl')

# Define a custom TensorFlow layer
class RandomForestLayer(tf.keras.layers.Layer):
    def __init__(self, rf_model):
        super(RandomForestLayer, self).__init__()
        self.rf_model = rf_model

    def call(self, inputs):
        # Convert inputs to NumPy for compatibility with scikit-learn model
        np_inputs = inputs.numpy()
        rf_probs = self.rf_model.predict_proba(np_inputs)
        return tf.convert_to_tensor(rf_probs, dtype=tf.float32)

# Define a function to make a Keras model that is functionally equivalent
def scikit_learn_to_keras(rf_model):
    inputs = tf.keras.Input(shape=(X_train.shape[1],))
    rf_layer = RandomForestLayer(rf_model)(inputs)
    
    # Create the equivalent Keras model
    keras_model = tf.keras.Model(inputs=inputs, outputs=rf_layer)
    return keras_model

# Convert scikit-learn Random Forest model to Keras model
tf_keras_model = scikit_learn_to_keras(loaded_rf_model)

# Save the Keras model as a TensorFlow SavedModel (using the new recommended format)
tf_keras_model.save('network_traffic_classifier_saved_model.keras')

print("Model has been successfully converted and saved as TensorFlow SavedModel format.")
```
